```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_excel('TTM.xlsx')
```

```python
dataset
```

```python

x = dataset.iloc[:, :-1].values
```

```python
y = dataset.iloc[:, -1].values
```

```python
y = y.reshape(len(y),1)
```

```python
from sklearn.svm import SVR
regressor = SVR()
regressor.fit(x, y)
```

```python
y_pred = regressor.predict(x)
```

```python
plt.scatter(x, y, color = 'pink')
plt.plot(x, regressor.predict(x), color = 'red')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'linear')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'black')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'poly')
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'pink')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```

```python
regressor = SVR(kernel = 'rbf',C=1.0, epsilon=0.2)
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y, color = 'silver')
plt.plot(x, regressor.predict(x), color = 'pink')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```
